```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
(r, c) = app_train.shape
print('row size: %d; column size: %d.' % (r, c))
```

#### 列出所有欄位

```python
print(app_train.columns)
```

#### 截取部分資料

```python
n = 5
# 看前 n 行
app_train.head(n)
```

```python
# 看後 n 行
app_train.tail(n)
```

```python
# 擷取 row index = 4 的資訊
app_train.iloc[4]
```

```python
# 擷取 column index = 5 的資訊
app_train.iloc[:, 5]
```

```python
# 擷取前 3 row 的資訊
app_train.iloc[:3]
```

```python
# 擷取前 3 row 和前 5 column 的資訊
app_train.iloc[:3, :5]
```

```python
# 用 column name 選取資訊
app_train.loc[1:3, ['TARGET', 'CODE_GENDER', 'AMT_INCOME_TOTAL']]
```
